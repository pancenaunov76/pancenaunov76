За секоја држава, најдете која година инфлацијата била највисока.

```python
countries = df['Country'].unique().tolist()
for country in countries:
  tmp = df[df['Country'] == country]
  tmp['Inflation'] = tmp['Inflation'].replace(" ",np.nan).replace('  ', np.nan)
  tmp['Inflation'] = tmp['Inflation'].fillna(0)
  tmp['Inflation'] = tmp['Inflation'].astype(float)
  try:
    year = tmp[tmp['Inflation'] == tmp['Inflation'].max()]['Year'].value_counts().index[0]
    print(country, "-", year)
  except:
    print('No information about', country)
```

Извадете ги државите кои имале Domestic Debt.

```python
df[df['Domestic Debt']>0]['Country'].unique()
```

Извадете ги државите кои имале Domestic Debt во 1945.

```python
df[(df['Domestic Debt']>0) & (df['Year']==1945.0)]['Country'].unique()
```

Кои години Унгарија имала Systemic Crisis?

```python
df[(df['Country']=='Hungary') & (df['Systemic Crisis'] > 0)]['Year'].astype(int).values
```

За секоја година извадете кои држави имале Currency Crises.

```python
unique_year = df['Year'].astype(int).unique()
for i in unique_year:
  a = df[(df['Year']==i) & (df['Currency Crises']==1)]['Country'].values
  if len(a) > 0:
    print(f"Vo {i} godina, Currency Crises imale: {a}.")
```

```python
jobs_df.info()
```

За секој department најдете која позиција е најзастапена.

```python
#jobs_df.groupby('Department')['Job Titles'].value_counts()
unique_department = jobs_df['Department'].unique()
#unique_department
for i in unique_department:
  most_freq_job = jobs_df[jobs_df['Department'] == i]['Job Titles'].value_counts().index[0]
  print(f"Vo departmentot {i}, najzastapena pozicija e {most_freq_job}.")
```

Извадете ги вработените кои имаат највисока плата.

```python
for i in range (len(jobs_df)):
  salary = str(jobs_df['Annual Salary'][i])
  #print(salary)
  if salary != 'nan' and salary.__contains__("$"):
    salary = salary.replace("$", "")
  salary = float(salary)
  jobs_df.at[i, 'Annual Salary'] = salary
jobs_df[jobs_df['Annual Salary'] == jobs_df['Annual Salary'].max()]
```

Извадете ги вработените кои имаат најниска дневница (по час).

```python
for i in range(len(jobs_df)):
  salary = str(jobs_df['Hourly Rate'][i])
  #print(salary)
  if salary != 'nan' and salary.__contains__("$"):
    salary = salary.replace("$", "")
  salary = float(salary)
  jobs_df.at[i, 'Hourly Rate'] = salary
jobs_df[jobs_df['Hourly Rate'] == jobs_df['Hourly Rate'].min()]
```

Извадете ги имињата на вработените кои работат Full time.

```python
jobs_df[jobs_df['Full or Part-Time']=='F']['Name']
```

За секоја професија најдете кој вработен зема највисока плата.

Hint: Треба да користите try/except за да се справиме со missing values

```python
unique_jobtitle = jobs_df['Job Titles'].unique()

for i in unique_jobtitle:
  tmp = jobs_df[(jobs_df['Job Titles'] == i)]
  try:
    name = tmp[tmp['Annual Salary'] == tmp['Annual Salary'].max()]['Name'].value_counts().index[0]
    print(i, "->", name)
  except:
    print('No information about', i)
```
